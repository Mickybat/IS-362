```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/dataflights/untidy/untidy_student_grades.csv')
df
```

### Reshaping the data
-I first decided to break the sex_and_age column into two columns as I thought it would be more appropiate to have those 2 values in two seperate columns. I did so by splitting the strings around the '_ ' separator, and also dropped the sex_and_age column by using drop() method as it was no longer needed.

```python
df[['sex', 'age']] = df['sex_and_age'].str.split('_', 1, expand=True) 
df = df.drop(columns=['sex_and_age'])
df = df[['id', 'name', 'phone', 'sex', 'age', 'test_number', 'term_1', 'term_2', 'term_3']] 
```

### Reshaping the data 
-I decided to tidy up the data by breaking the table into two tables, one containing test 1 and another containing test 2.

```python
group = df.groupby('test_number')  
test_1 = group.get_group('test 1')  
test_2 = group.get_group('test 2')  
```

### Reshaping the data
-To do calculations, I decided to merge the two tables by 'id' in order to have one row with all the information from each student.

```python
all_test = test_1.merge(test_2, on='id') 
all_test.drop(['name_y', 'sex_y', 'phone_y', 'age_y'], axis = 1)
```

### Calculations 
-This calculation gets the average score of all the 'test 1' tests.

```python
terms = all_test[['term_1_x', 'term_2_x', 'term_3_x']].mean()
terms.mean()
```
